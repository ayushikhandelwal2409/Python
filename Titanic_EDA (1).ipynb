EDA on Titatic dataset



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import scipy

# Load the Titanic dataset
data=sns.load_dataset('titanic')
```

```python
# First 5 rows
data.head()
```

```python
# Dataset info
data.info()
```

```python
# Summary statistics
data.describe(include='all')
```

```python
# Check for missing values
data.isna().sum()
```

```python
data['age'] = data['age'].fillna(data['age'].median())

# Drop row with missing 'embarked'
data.dropna(subset=['embarked'], inplace=True)
```

```python
# Removing row with missing values
data.drop('deck', axis=1, inplace=True)
```

```python
data[data.duplicated()].shape
```

```python
# Removing duplicates
data.drop_duplicates(inplace=True)
```

```python
data[data.duplicated()].shape
```

Univariate Analysis
Categorical Features:

```python
# Countplots for categorical variables
sns.countplot(data,x='sex')
plt.title('Gender Distribution')
plt.show()

sns.countplot(x='class', data=data)
plt.title('Passenger Class Distribution')
plt.show()

sns.countplot(x='embarked', data=data)
plt.title('Port of Embarkation')
plt.show()

sns.countplot(x='who', data=data)
plt.title('Who were on board')
plt.show()
```

Numerical Features

```python
# Histogram for age
plt.hist(data['age'], bins=30)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.grid()
plt.show()

# Fare distribution
sns.histplot(data,x='fare', kde=True)
plt.title('Fare Distribution')
plt.show()
```

Bivariate Analysis
Survival Rate by Category:
<!--  -->here 1 means alive and 0 means died

```python
# Survival by gender
sns.countplot(x='sex', hue='survived', data=data)
plt.title('Survival by Gender')
plt.show()

# Survival by class
sns.countplot(x='class', hue='survived', data=data)
plt.title('Survival by Class')
plt.show()

# Survival by embarkation
sns.countplot(x='embarked', hue='survived', data=data)
plt.title('Survival by Embarked Port')
plt.show()
```

Multivariate Analysis
Correlation Heatmap

```python
# Numeric correlation matrix
corr = data.corr(numeric_only=True)
sns.heatmap(corr, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(data, hue='sex', x_vars=['age', 'fare'], y_vars=['sibsp', 'parch'])
plt.show()
```
